## loc

```python
# Intervalo entre as linhas 3 e 5
df.loc[3:5]
```

```python
# Intervalo das linhas 7 a 10
df.loc[7:10]
```

```python
# Exibir linhas específicas (2, 8, 20)
df.loc[[2, 8, 20]]
```

```python
# Filtrar dados com uma condição específica (where)
# Dados em que o tipo de programa seja TV Show
df.loc[df['type'] == 'TV Show']
```

```python
# Filtrar os dados em que o diretor é Steven Spielberg
df.loc[df['director'] == 'Steven Spielberg']
```

```python
# Exibir os filmes entre 2015 e 2020
df.loc[(df['release_year'] > 2015) & (df['release_year'] < 2020)]
```

```python
# Criar um novo dataframe, a partir dos dados filtrados
df2 = df.loc[df['type'] == 'Movie']
print(df2)
```

```python
df3 = df.loc[df['type'] == 'TV Show']
print(df3)
```

```python
# Exibir titulo, diretor e elenco nos filmes lançados após 2015
df4 = df[['title', 'director', 'cast']].loc[df['release_year'] > 2015]
print(df4)
```

## Renomear colunas

```python
df.rename(columns = {
    'title': 'titulo',
    'director': 'diretor',
    'type': 'tipo'
})
```

```python
df.head(2)
```

```python
# Para renomear de forma permanente:
df.rename(columns = {
    'title': 'titulo',
    'director': 'diretor',
    'type': 'tipo'
}, inplace = True)
```

```python
df.head(1)
```

```python
# Renomear country = pais, duration = duracao, cast = elenco
df.rename(columns = {
    'country': 'pais',
    'duration': 'duracao',
    'cast': 'elenco'
}, inplace = True)
```
